```python
import pandas as pd
```

```python
df = pd.read_excel('../data/population.xlsx', header=0)
df = df.reset_index(drop=True).T.reset_index().T

df.loc[df.iloc[:,0] == 'Germany (until 1990 former territory of the FRG)', 0] = 'Germany'
df.columns = ['country', 'population']
df = df.set_index('country')
print(df.shape)
df.head()
```

```python
da = pd.read_excel('../data/happiness.xlsx', header=0, index_col='country')
countries = list(da.index)
countries
```

```python
df = df.loc[countries,:]
print(df.shape)
df.head()
```

```python
for feat in ['happiness', 'bad_conditions', 'culture', 'health_estimate']:
    
    da = pd.read_excel(f'../data/{feat}.xlsx', header=0, index_col='country')    
    da = da.loc[countries,:]    
    df[feat] = da.iloc[0:]
    
df
```

```python
for feat in ['commute', 'education_underachievements', 'minimum_wage', 'unmet_medical_needs']:
    
    da = pd.read_excel(f'../data/{feat}.xlsx', header=0)
    
    da = da.reset_index(drop=True).T.reset_index().T

    da.loc[da.iloc[:,0] == 'Germany (until 1990 former territory of the FRG)', 0] = 'Germany'
    da.columns = ['country', feat]
    
    
    da = da.set_index('country')    
    da = da.loc[countries,:]    
    df[feat] = da.iloc[0:]
```

```python
df.to_excel('../data/main.xlsx')
```
