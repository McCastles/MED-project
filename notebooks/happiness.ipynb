```python
import pandas as pd
```

```python
org = pd.read_csv('../eurostat/raw_happiness.csv')
```

```python
org.columns
```

```python
df = org[['LEV_SATIS', 'GEO', 'Value']]
df.columns = ['level', 'country', 'value']
df = df[~df.country.str.contains('European')]
df.loc[df['country'] == 'Germany (until 1990 former territory of the FRG)', 'country'] = 'Germany'
df = df.reset_index(drop=True)
# list(df.country)
df
```

```python
def calculate_mean_happiness(countries):
    a = df[df['country'].isin(countries)].drop(['country'],axis=1) 
    a.value = a.value.astype(float)
    return a.groupby(['level']).mean().value
```

```python
def colons(country, a):
    print(a)
    for i, level in enumerate(['High', 'Low', 'Medium']):
        
#         print(i, level, country)
        condition = (df['country'] == country) & (df['level'] == level)
#         print('\n', df.loc[condition, :])
        df.loc[condition, 'value'] = a[i]
    return df[df['country'] == country]
```

```python
colons('Turkey', calculate_mean_happiness(['Greece', 'Croatia']))
```

```python
colons('Iceland', calculate_mean_happiness(['Norway', 'Denmark', 'Nehterlands', 'Findland']))
```
