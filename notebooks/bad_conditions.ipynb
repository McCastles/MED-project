```python
import pandas as pd
```

```python
org = pd.read_csv('../eurostat/raw_bad_conditions.csv')
```

```python
org.columns
```

```python
org
```

```python
df = org[ (org['TIME'] == 2018) ][['GEO', 'Value']]
df.columns = ['country', 'value']
df.loc[df['country'] == 'Germany (until 1990 former territory of the FRG)', 'country'] = 'Germany'
df = df.sort_values('country')
print(df.shape)
df.head()
# df
```

```python
df[df['value'] == ':']['country']
```

```python
for c in df[df['value'] == ':']['country']:

    v = org[ (org['TIME'] == 2016) & (org['GEO'] == c) ]['Value']
    
    df.loc[df['country'] == c, 'value'] = float(v)
    
df[df['value'] == ':']
```

```python
df = df.set_index('country')
df.columns=['health']
```
