```python
import pandas as pd
```

```python
org = pd.read_csv('../eurostat/raw_health_estimate.csv')
```

```python
org.columns
```

```python
org.head()
```

```python
df = org[ (org['TIME'] == 2018) & (org['LEV_PERC'] == 'Very good or good')][['GEO', 'Value']]
df.columns = ['country', 'value']
# df = df[df['health'] == '']
df.loc[df['country'] == 'Germany (until 1990 former territory of the FRG)', 'country'] = 'Germany'
df = df.sort_values('country')
df.head()
df
```

```python
df[df['value'] == ':']['country']
```

```python
for c in df[df['value'] == ':']['country']:

    v = org[ (org['TIME'] == 2016) & (org['GEO'] == c) & (org['LEV_PERC'] == 'Very good or good')]['Value']
    
    df.loc[df['country'] == c, 'value'] = float(v)
    
df[df['value'] == ':']
```

```python
df = df.set_index('country')
df.columns=['health']
```
